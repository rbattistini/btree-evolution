```kotlin
%use lets-plot
%use dataframe
```

```kotlin
import java.util.Locale
val csvPath = "./simulation_results.csv"
val csv = DataFrame.readCSV(csvPath, parserOptions = ParserOptions(locale = Locale.UK))
```

```kotlin
val data = csv.select { "btree" and "fitness" }
```

```kotlin
val boxPlot = ggplot(data.toMap()) { x = "btree"; y = "fitness" } +
  ggsize(1200, 600) +
  ggtitle("Test") +
  geomBoxplot(whiskerWidth = 0.2)
```

```kotlin
boxPlot
```

```kotlin
ggsave(boxPlot, "btrees_evaluation.png")
```
